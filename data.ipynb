```python
import pandas as pd
```

```python
df_iyong = pd.read_csv('data/iyong_7_12.csv',encoding='CP949')
df_iyong.columns = ['date', 'n', 'code', 'sex', 'age', 'use', 'moment', 'carbon', 'distance', 'time']
df_iyong.head()
```

```python
# 이동거리 0이면 제거
# 성별?
```

```python
idx = df_iyong[df_iyong['distance'] == 0].index
df_iyong.drop(idx , inplace=True)
```

```python
df_iyong['sex'] = df_iyong['sex'].fillna('N')
```

```python
df_iyong.isnull().sum()
```

```python
df_iyong.dtypes
```

```python
df_iyong.loc[df_iyong['sex'] == 'm', 'sex'] = 'M'
df_iyong.loc[df_iyong['sex'] == 'f', 'sex'] = 'F'
```

```python
#df_iyong['code'].value_counts()
#df_iyong['sex'].value_counts()
df_iyong['age'].value_counts()
```

```python
df_iyong.describe(include='all')
```

```python
df_iyong = df_iyong.astype({'moment':'float', 'carbon':'float'})
```

```python
df_iyong_dummy = pd.get_dummies(df_iyong)
```

```python
df_iyong_dummy.columns
```

```python
df_num = df_iyong_dummy.groupby(["date","n"]).agg({"use":"sum","moment":"mean", "carbon":"mean", "distance":"mean", 'time':'mean'}).round(2)
df_num = df_num.reset_index()
df_num
```

```python
df_dummy = df_iyong_dummy[['date', 'n', 'code_단체권',
       'code_일일권', 'code_일일권(비회원)', 'code_정기권', 'sex_F', 'sex_M', 'sex_N',
       'age_20대', 'age_30대', 'age_40대', 'age_50대', 'age_60대', 'age_70대이상',
       'age_~10대', 'age_기타']]
df_dum = df_dummy.groupby(["date","n"]).agg(sum).reset_index()
df_dum
```

```python
df_new_iyong = pd.merge(df_num, df_dum, on=['date', 'n'])
df_new_iyong
```

```python
df_jangso = pd.read_csv('data/jangso_7_12.csv',encoding='CP949')
df_jangso.columns = ['gu', 'n', 'date', 'rent', 'return', 'in_rate', 'out_rate']
df_jangso
#건수가 -로 누락된 행 5개 제거
```

```python
df_all = pd.merge(df_jangso, df_new_iyong, how='inner', on=['date', 'n'])
df_all
```

```python
df_all[df_all.isna().any(axis=1)] #9개 데이터 제외
```

```python
df_all.to_csv('data/all_7_12.csv', index = False, encoding='utf-8')
```
